```python
import numpy as np
import matplotlib.pyplot as plt
from get_data import fig_data,xor_data,donut_data

############################  using cost function batch gradient descent

def get__data():

    w=np.array([-1.5,1.5])
    x=np.random.random((1000,2))*2-1
    b=0.1
    #y=np.dot((w.T),x)
    y=np.sign(np.dot(x,w)+b)
    x[y==1.0]=x[y==1.0]-np.array([-0.1,0])
    x[y==-1.0]=x[y==-1.0]+np.array([+0.1,0])
    return x,y

class perceptrons:
        
    
    def fit(self,x,y,learning_rate=1.0,epochs=10000):
        
        D=np.shape(x)[1]
        N=np.shape(x)[0]
        self.w=np.random.randn(D)
        self.b=0
        costs=[]
        
        
        for i in range(epochs):
            one_func=np.zeros(len(y))
            Y_pred=self.predict(x)
            
            list_false=np.nonzero(Y_pred!=y)[0]
            
            for k in range(len(y)):
                if (Y_pred[k]!=y[k]):
                    one_func[k]=1.0
                    
            
            if len(list_false)==0:
                
                break
            
 
            #j=np.random.choice(list_false)
            #self.w=self.w+learning_rate*y[j]*x[j]
            #self.b=self.b+learning_rate*y[j]*1
            
            # use cost function
            self.w=self.w+learning_rate*np.dot((x.T),y*one_func)
            self.b=self.b+learning_rate*np.sum(y*one_func)
            
            accu=len(list_false)/float(N)
            costs.append(accu)
                
        print('w,b,iteration',self.w,self.b,i)    
        plt.plot(costs) 
        plt.show()
            
    def predict(self,x):
        
        return np.sign(np.dot(x,self.w)+self.b)
    
    def score(self,x,y):
        Y_pred=self.predict(x)
        plt.scatter(x[:,0],x[:,1],c=Y_pred,alpha=0.5)
        plt.show()
        score_=np.mean(Y_pred==y)
        return score_
        
            
                
if __name__=='__main__':
    
    Xtrain,Ytrain=get__data()
    plt.scatter(Xtrain[:,0],Xtrain[:,1],c=Ytrain,alpha=0.5)
    plt.show()
    
    #Ntrain=int(len(y)/2)
    #Xtrain=x[:Ntrain]
    #Ytrain=y[:Ntrain]
                   
    #Xtest=x[Ntrain:]
    #Ytest=y[Ntrain:]
    Xtest,Ytest=get__data()    
    
    percep=perceptrons()
    percep.fit(Xtrain,Ytrain)
    print('Train score:',percep.score(Xtrain,Ytrain))
    print('Test score:',percep.score(Xtest,Ytest))
    
    
```
