```python
import numpy as np
import matplotlib.pyplot as plt
from get_data import fig_data,xor_data,donut_data


def get__data():

    x=np.zeros((1000,2))
    y=np.zeros(1000)
    #X[0:50,:]=np.random.random((50,2))/2.0
    #X[50:100,:]=np.random.random((50,2))/2.0+0.5
    #X[100:150,:]=np.random.random((50,2))/2.0+np.array([0.5,0])
    #X[150:200,:]=np.random.random((50,2))/2.0+np.array([0.,0.5])   
    #Y[0:200]=np.array(100*[0]+100*[1])
    
    
    x[0:500,:]=np.random.random((500,2))*2+.9
    x[500:1000,:]=np.random.random((500,2))*2-.9

    y[0:1000]=np.array(500*[-1.0]+500*[1.0])
    
    return x,y

class perceptrons:
        
    
    def fit(self,x,y,learning_rate=1.0,epochs=10000):
        
        D=np.shape(x)[1]
        N=np.shape(x)[0]
        self.w=np.random.randn(D)
        self.b=0
        costs=[]
        
        for i in range(epochs):
            
            Y_pred=self.predict(x)
            
            list_false=np.nonzero(Y_pred!=y)[0]
            
            if len(list_false)==0:
                
                break
            
            #for j in list_false:       #all the content change at the time, does not work. one by one works
                
            #    self.w=self.w+learning_rate*y[j]*x[j]
            #    self.b=self.b+learning_rate**y[j]*1
            #choose one sample 
            j=np.random.choice(list_false)
            self.w=self.w+learning_rate*y[j]*x[j]
            self.b=self.b+learning_rate*y[j]*1
            
            accu=len(list_false)/float(N)
            costs.append(accu)
                
        print('w,b,iteration',self.w,self.b,i)    
        plt.plot(costs)    
        plt.show()
            
    def predict(self,x):
        
        return np.sign(np.dot(x,self.w)+self.b)
    
    def score(self,x,y):
        Y_pred=self.predict(x)
        score_=np.mean(Y_pred==y)
        plt.scatter(x[:,0],x[:,1],c=Y_pred,alpha=0.5)
        plt.show()
        return score_
        
            
                
if __name__=='__main__':
    
    Xtrain,Ytrain=get__data()
    plt.scatter(Xtrain[:,0],Xtrain[:,1],c=y,alpha=0.5)
    plt.show()
    
    Xtest,Ytest=get__data()
                   
    percep=perceptrons()
    percep.fit(Xtrain,Ytrain)
    print('Train score:',percep.score(Xtrain,Ytrain))
    print('Test score:',percep.score(Xtest,Ytest))
    #x,y=get__data()

    
    
```
